***CARGAR DATOS***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/scanavi31/03MIAR--Python_para_IA/main/BankChurners.csv'
df = pd.read_csv(url)
# Ver las primeras 5 filas
print(df.head())
```

***Explorar los datos***

```python
print("Información de las primeras 5 filas: \n")
print(df.head())  # devuelve las primeras n filas de un objeto DataFrame. Por defecto, n es igual a 5
print(" \n Información descriptva del DataFrame: \n ")
print(df.describe()) #proporciona estadísticas descriptivas sobre las columnas numéricas de un objeto DataFrame.
print(" \n Información básica del DataFrame: \n ")
print(df.info()) # muestra información básica sobre un objeto DataFrame
```

***Realiza un análisis estadístico básico de los datos***

```python
# Contar el número de clientes que se han ido y se han quedado
print(df["Attrition_Flag"].value_counts())
print("\n")
# Calcular la edad promedio de los clientes
print("Edad promedio de los clientes:", round(df["Customer_Age"].mean()))
print("\n")
# Calcular la relación entre los ingresos y la tasa de utilización del crédito
credit_utilization = df["Avg_Utilization_Ratio"].groupby(df["Income_Category"]).mean()
print(credit_utilization)
```

***Crea gráficos para visualizar los datos***

```python
# Histograma de la edad de los clientes
plt.hist(df["Customer_Age"], bins=20)
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
# Agregar título
plt.title("Histograma de la edad de los clientes")
plt.show()

# Gráfico de barras de los ingresos de los clientes
sns.countplot(x="Income_Category", data=df)
plt.xticks(rotation=45)
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.title("Gráfico de barras de los ingresos de los clientes")
plt.show()

# Gráfico de dispersión de los ingresos y la tasa de utilización del crédito
plt.scatter(df["Credit_Limit"], df["Avg_Utilization_Ratio"])
plt.xlabel("Límite de crédito")
plt.ylabel("Tasa de utilización del crédito")
plt.title("Gráfico de dispersión de los ingresos y la tasa de utilización del crédito")
plt.show()
```

***Análisis Objetivo Planteado***

```python
# Ver la distribución de los datos demográficos
print(df['Gender'].value_counts(),"\n")
print(df['Marital_Status'].value_counts(),"\n")
print(df['Education_Level'].value_counts(),"\n")
print(df['Income_Category'].value_counts(),"\n")

# Ver la tasa de abandono
print(df['Attrition_Flag'].value_counts(normalize=True))
```

```python
# Calcular la tasa de abandono de los clientes
attrition_rate = df["Attrition_Flag"].value_counts(normalize=True) * 100
print("Tasa de abandono de los clientes:")
print(attrition_rate)

# Visualizar la tasa de abandono de los clientes
plt.figure(figsize=(8,6))
sns.countplot(x="Attrition_Flag", data=df)
plt.title("Tasa de abandono de los clientes")
plt.xlabel("Abandono")
plt.ylabel("Número de clientes")
plt.show()

# Analizar la relación entre la tasa de abandono y los datos demográficos
plt.figure(figsize=(10,8))
sns.countplot(x="Attrition_Flag", hue="Gender", data=df)
plt.title("Tasa de abandono por género")
plt.xlabel("Abandono")
plt.ylabel("Número de clientes")
plt.show()

plt.figure(figsize=(10,8))
sns.countplot(x="Attrition_Flag", hue="Education_Level", data=df)
plt.title("Tasa de abandono por nivel educativo")
plt.xlabel("Abandono")
plt.ylabel("Número de clientes")
plt.show()

plt.figure(figsize=(10,8))
sns.countplot(x="Attrition_Flag", hue="Income_Category", data=df)
plt.title("Tasa de abandono por categoría de ingresos")
plt.xlabel("Abandono")
plt.ylabel("Número de clientes")
plt.show()

plt.figure(figsize=(10,8))
sns.countplot(x="Attrition_Flag", hue="Marital_Status", data=df)
plt.title("Tasa de abandono por estado civil")
plt.xlabel("Abandono")
plt.ylabel("Número de clientes")
plt.show()
```
